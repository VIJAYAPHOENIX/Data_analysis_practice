<h2 style = 'color : orange'>GROUPBY</h2>
-> It groups the values which are connected to each other.<br>
--> Groupby process:<br>
        1. Data acquire.<br>
        2. Splitting the data.<br>
        3. Appling operations(Aggregate operations).<br>
        4. Combined results<br>

```python
p = ["Samsung", "Nokia", "Samsung", "Motorola", "Nokia", "Samsung", "Samsung"]
o = [2,4,3,4,6,7,3]

d = {
    "Product": p,
    "Orders":o
}

df1 = pd.DataFrame(d)
df1
```

```python
g = df1.groupby("Product")
print(g.sum())
print(g.mean())
```

```python
invt = stp_df.groupby("InvestmentnType")
invt.size()
```

```python
col = ["City  Location", "InvestmentnType"]           #IT gives output as Series
a = stp_df.groupby(col )["City  Location"]
b = a.sum()
print(b)
```

```python
col = ["City  Location", "InvestmentnType"]           #IT gives output as DataFrame by using "as_index = False"
a = stp_df.groupby(col, as_index = False)["City  Location"]
b = a.sum()
print(b)
```

<h2>MERGING AND JOINING</h2>
-->pd.merge(p1,p2,p3,p4) is used to merge two dataframes.<br>
1. Here "p1" parameter refers to dataframe1.<br>
2. Here "p2" parameter refers to dataframe2.<br>
3. Here "p3" parameter refers to common column in both dataframes.<br>
4. Here "p4" parameter refers to type of join or merging we are going to use.<br>

```python
d1 = {
    "Id":[1, 2, 3, 4, 5, 6],
    "Name":["Prasad", "Venu", "Madhu", "Naresh", "Sai", "Veeru"],
    "Subject":["English", "Java", "Html", "Python", "C", "Dot net"]
}

d2 = {
    "Id":[11, 12, 13, 14, 15, 16],
    "Name":["Srinu", "Sumanth", "Neel", "Daniel", "Arjun", "Veeru"],
    "Subject":["Java", "Html", "Cpp", "Python", "C", "Dot net"]
}

df1 = DataFrame(d1)
df2 = DataFrame(d2)

print(df1)
print(df2)
```

```python
inner_join = pd.merge(df1, df2, on = "Subject", how = "inner")
inner_join
```

```python
left_join = pd.merge(df1, df2, on = "Subject", how = "left")
left_join
```

```python
right_join = pd.merge(df1, df2, on = "Subject", how = "right")
right_join
```

```python
outer_join = pd.merge(df1, df2, on = "Subject", how = "outer")
outer_join
```

 <h2 style = "color : orange">Pandas - DataFrame - Data time</h2>
 --> We use "parse_dates = [column_name]" when we import the data/file.<br> 
 --> suppose, If it is not working then we have to do it manually.<br>
 --> "df[column_name] = pd.to_datetime(df[column_name], format = '%d%m%Y / %m%d%Y / %Y%m%d')" is used after loading the data.
 If it has a string value such as 12sep2019 we have to use format = '%d%b%Y.'<br>
 --> If we have any irrelevent errors then we use errors = 'coerce' then it converts the errors as "NaT"(Not A Time).<br>

```python
sdf = pd.read_csv( "D:\Datasets\startup_funding.csv",
                  parse_dates = ['Date dd/mm/yyyy'])
modified_col = {'Sr No': 'Sl_no',
                'Date dd/mm/yyyy':'Date',
                'Startup Name':'Startup_name',
                'Industry Vertical':'Industry_Vertical',
                'City  Location':'Location',
                'Investors Name':'Investors',
                'InvestmentnType':'Fund_Type',
                'Amount in USD':'Funding'}
sdf.rename(columns = modified_col, inplace = True)
```

```python
sdf.columns
```

```python
sdf["Date"] = sdf["Date"].str.replace("\xa0", "", regex = False).str.strip()     #After cleaning date column we convert it manually.

sdf["Date"] = pd.to_datetime(sdf["Date"], dayfirst = True, errors = "coerce")
sdf.info()
```

```python
import os 
import glob
path = "D:\Datasets\monthly_sales"                       #By using "glob" module we are creating the seperate paths to files
paths = os.path.join(path,"*.csv")
files = glob.glob(paths)
files
```
