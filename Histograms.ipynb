```python
# standard imports
import numpy as np
import pandas as pd
from numpy.random import randn
```

```python
# stats
from scipy import stats
```

```python
#plotting
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
dataset1 = randn(100)
plt.hist(dataset1,bins=20,edgecolor='black')
plt.grid(True,linestyle='--',alpha=0.5)
```

```python
dataset2 = randn(80)
plt.hist(dataset2,color='indianred)
```

```python
plt.hist(dataset1, density=True, color='indianred', alpha=0.5, bins=20)
plt.hist(dataset2,density=True,alpha=0.5,bins=20)
plt.grid(True, linestyle='--', alpha=0.7)
```

```python
data1 = randn(1000)
data2 = randn(1000)
```

```python
df1 = pd.DataFrame({'x':data1,'y':data2})
sns.jointplot(x='x',y='y',data=df1,kind='scatter')
```

```python
sns.jointplot(x='x',y='y',data=df1,kind='hex')
```

```python
sns.jointplot(x='x',y='y',data=df1,kind='kde')
```

```python
sns.jointplot(x='x',y='y',data=df1,kind='reg')
```
