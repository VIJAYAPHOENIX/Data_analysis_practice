<h2 style = 'color: orange'>Sampling of data</h2>
<h3 style = 'color: blue'>Types of Sampling :</h3>
<h4 style = 'color: green'>1. Sampling without Replacement</h4>
    Here we take samples of data from the dataset either Orderly or Randomly. In this their is no repeating of data.
<h4 style = 'color: green'>2. Sampling with Replacement</h4>
    Here we take samples of data from the dataset either Orderly or Randomly. But in this we can get values repeatedly. Here it is called as "Boot Strapping" is often used in machine learning models when the data is limited.
<h4 style = 'color: green'>3. Stratified Sampling</h4>
    It is used to divide the data into stratified group based on certain keys such as gender, population.. . It takes the 1/3 data to the sampling.

<h2>Prediction Accuracy :</h2>
    Predictive accuracy measures how many predictions were correct out of the total test samples.
$$
\text{Predictive Accuracy} = \frac{\text{correct predictions}}{\text{ Number of Test Instances}}
$$
<h2>Mean Absolute Error :</h2>
    Mean absolute error measures the average absolute difference between predicted and actual values.
$$
\text{MEA} = \frac{\sum|\text{Predicted} - \text{Actual}|}{\text{Number of Test Instances}}
$$    

<h2 style = 'color: orange'>How to Build Machine Learning Model in Python</h2>
    It is explained in the following steps.

<h3>1. Collect The Data</h3>

```python
bikes = pd.read_csv(r"D:\Datasets\ML_basics\bikes.csv")
bikes.head()
```

<h3>2. Explore The Data</h3>

```python
bikes.info()
```

```python
bikes.describe()
```

```python
bikes.plot(x= 'temperature', y= 'rentals', kind = 'scatter')
```

```python
bikes.plot(x= 'humidity', y= 'rentals', kind = 'scatter')
```

```python
bikes.plot(x= 'windspeed', y= 'rentals', kind = 'scatter')
```

<h3>3. Prepare The Data</h3>

```python
response = 'rentals'
y = bikes[[response]]
y
```

```python
predictors = list(bikes.columns)
predictors.remove(response)
x = bikes[predictors]
x
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state = 1234)
```

<h3>4. Train The Model</h3>

```python
from sklearn.linear_model import LinearRegression as lr
model = lr().fit(x_train, y_train)
```

```python
model.intercept_
```

```python
model.coef_
```

<h3>5. Evaluate The Model</h3>

```python
model.score(x_test, y_test)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_test, y_pred)
```
