```python
# Normal imports
import numpy as np
import pandas as pd
from numpy.random import randn

# Stats
from scipy import stats

import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
dataset = randn(100)
```

```python
sns.displot(dataset)
```

```python
sns.histplot(dataset, bins=25, stat='density',
            color='indianred', label='HIST')
sns.kdeplot(dataset,
            color='blue', label='KDE PLOT')
```

```python
# using series as data using pandas
ser1 = pd.Series(dataset,name='my data')
sns.histplot(ser1, bins=25, stat='density',
            color='indianred', label='HIST')
sns.kdeplot(ser1,
            color='blue', label='KDE PLOT')
```
