```python
# normal imports
import numpy as np
import pandas as pd
from numpy.random import randn

# stats
from scipy import stats

# visualisation libraries
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
# random data
data1 = randn(100)
data2 = randn(100)

sns.boxplot([data1,data2])
```

```python
sns.boxplot([data1,data2],whis=np.inf)
```

```python
sns.boxplot(data1,vert=False)
```

```python
sns.boxplot(data1,orient='h',whis=np.inf)
```

```python
# creating new data to demostrate both boxplot and violinplot
# normal dist
data1 = stats.norm(0,5).rvs(100)

# Two gamma dist. concatenate together
data2 = np.concatenate([stats.gamma(5).rvs(50)-1,
                       -1*stats.gamma(5).rvs(50)])

sns.boxplot([data1,data2],whis=np.inf)
```

```python
sns.violinplot([data1,data2])
```

```python
sns.violinplot([data1,data2],bw_method=0.01)
```

```python
sns.violinplot(data1,inner='stick')
```
