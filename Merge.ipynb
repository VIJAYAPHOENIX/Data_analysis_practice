```python
import numpy as np
import pandas as pd 
from pandas import Series,DataFrame
```

```python
df = DataFrame
```

```python
dframe1 = df({'key':['X','Z','Y','Z','X','X'],'data_set_1':np.arange(6)})
dframe1
```

```python
dframe2 = df({'key':['Q','Y','Z'],'data_set_2':[1,2,3]})
dframe2
```

```python
pd.merge(dframe1,dframe2)
```

```python
pd.merge(dframe1,dframe2,on='key',how='left')
```

```python
pd.merge(dframe1,dframe2,on='key',how='outer')
```

```python
dframe3 = df({'key':['X','X','X','Y','Z','Z'],'data_set_3':range(6)})
dframe3
```

```python
dframe4 = df({'key':['Y','Y','X','X','Z'],'data_set_4':range(5)})
dframe4
```

```python
pd.merge(dframe3,dframe4,on='key',how='outer')
```

```python
# it is called chain merge calls
full_set=pd.merge(pd.merge(pd.merge(dframe1,dframe2,on='key',how='outer'),dframe3,on='key',how='outer'),dframe4,on='key',how='outer')
full_set
```

```python
# second solution
full_set2 = pd.concat([dframe1,dframe2,dframe3,dframe4],axis=0,join='outer')
full_set2
```
