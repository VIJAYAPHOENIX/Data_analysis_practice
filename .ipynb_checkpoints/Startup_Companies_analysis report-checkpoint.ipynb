```python
import pandas as pd
import numpy as np
from pandas import DataFrame, Series

from scipy import stats

import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams['figure.figsize'] = (10,6)
sns.set_style("whitegrid")
```

```python
file = "D:\Datasets\startup_funding.csv"
fund_df = pd.read_csv(file)
fund_df.head()
```

<h3>Data Cleaning</h3>

```python
fund_df.columns
```

```python
modified_col = {'Sr No': 'Sl_no',
                'Date dd/mm/yyyy':'Date',
                'Startup Name':'Startup_name',
                'Industry Vertical':'Industry_Vertical',
                'City  Location':'Location',
                'Investors Name':'Investors',
                'InvestmentnType':'Fund_Type',
                'Amount in USD':'Funding'}
fund_df.rename(columns = modified_col, inplace = True)
fund_df.columns
```

```python
fund_df["Funding"].isna().sum()
```

```python
fund_df['Funding'] = fund_df['Funding'].str.replace(",","")
```

```python
fund_df['Funding'] = pd.to_numeric(fund_df['Funding'],errors = 'coerce')
fund_df['Funding']
```

```python
fund_df['Funding'].skew()
```

```python
fund_df['Funding'] = fund_df['Funding'].fillna(fund_df['Funding'].median())
```

```python
fund_df['Log_Funding'] = np.log1p(fund_df['Funding'])
fund_df['Log_Funding'].skew()
```
