```python
# normal imports
import numpy as np
import pandas as pd
from numpy.random import randn

# stats
from scipy import stats

# visualization libraries
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams['figure.figsize'] = (10,6)
sns.set_style("whitegrid")

%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
print(tips.head())
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips)
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips,
          scatter_kws={'marker':'o', 'color':'indianred'},
          line_kws={'linewidth':1, 'color':'blue'})
plt.grid(True, linestyle='--', alpha=0.5)
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, order = 4,
          scatter_kws={'marker':'o', 'color':'indianred'},
          line_kws={'linewidth':1, 'color':'blue'})
plt.grid(True, linestyle='--', alpha=0.5)
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, fit_reg=False,
          scatter_kws={'marker':'o', 'color':'indianred'})
plt.grid(True, linestyle='--', alpha=0.5)
```

```python
tips['tip_perc'] = 100*(tips['tip']/tips['total_bill'])

tips.head()
```

```python
sns.lmplot(x='size', y='tip_perc', data=tips)
```

```python
sns.lmplot(x='size', y='tip_perc', data=tips, fit_reg=False)
```

```python
sns.lmplot(x='size', y='tip_perc', data=tips, x_jitter=.1)
```

```python
sns.lmplot(x='size', y='tip_perc', data=tips, x_estimator=np.mean)
```

```python
sns.lmplot(x='total_bill', y='tip_perc', data=tips, hue='sex', markers=['x','o'])
```

```python
sns.lmplot(x='total_bill', y='tip_perc', data=tips, hue='day')
```

```python
sns.lmplot(x='total_bill', y='tip_perc', data=tips, lowess=True, line_kws={'color':'Black'})
plt.grid(True,linestyle='--', alpha=0.6)
```

```python
sns.regplot(x='total_bill', y='tip_perc', data=tips)
```

```python
fig, (axis1, axis2) = plt.subplots(1,2,sharey=True)

sns.regplot(x='total_bill', y='tip_perc', data=tips, ax=axis1)
sns.violinplot(y='tip_perc', x='size', data=tips, color='red', ax=axis2)
```
